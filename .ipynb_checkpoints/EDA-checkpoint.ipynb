```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
plt.style.use("ggplot")
rcParams['figure.figsize'] = (12,  6)
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Class'] = df['Class'].map({'Business': 2, 'Eco Plus': 1, 'Eco':0})
df['Class'].value_counts()
```

```python
df['Customer Type'] = df['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer':0})
df['Customer Type'].value_counts()
```

```python
df['satisfaction'] = df['satisfaction'].map({'satisfied': 1, 'neutral or dissatisfied':0})
df['satisfaction'].value_counts()
```

```python
df['Type of Travel'] = df['Type of Travel'].map({'Business travel': 1, 'Personal Travel':0})
df['Type of Travel'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female':0})
df['Gender'].value_counts()
```

```python
df.describe()
```

# Categorical Variables

```python
def make_bars(data, name):
    data.plot(kind="bar")
    plt.title("Value counts of {}".format(name))
    plt.xlabel("satisfaction")
    plt.ylabel("Count")
    plt.show()
```

```python
df['satisfaction'].value_counts(normalize = True)
```

```python
make_bars(df['satisfaction'].value_counts(), 'Overall Satisfaction')
```

# Feature Relationships

```python
sns.set(rc={'figure.figsize':(20,16)})
hm = sns.heatmap(df.corr(), 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 cmap="Spectral_r")
plt.show()
```

```python
sns.catplot(x="satisfaction", y="Online boarding", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for target vs Online Boarding Satisfaction")
plt.show()
```
